```python
import pandas as pd
```

```python
# Import dataset
cleaned_decision_scores = pd.read_csv('cleaned_talent_decision_scores.csv', sep=',', header='infer')
cleaned_decision_scores.head()
```

```python
df = cleaned_decision_scores
```

### Strengths

```python
# Get all strengths into a list
all_strengths = []
for row in df['strengths'].dropna():
    all_strengths.extend([s.strip() for s in row.split(',')])
```

```python
# make unique strengths with IDs
strengths_df = pd.DataFrame({
    "strength_id": range(1, len(set(all_strengths)) + 1),
    "strength_name": list(set(all_strengths))
})
```

```python
strengths_df
```

### Weaknesses

```python
# Get all weaknesses into a list
all_weaknesses = []
for row in df['weaknesses'].dropna():
    all_weaknesses.extend([s.strip() for s in row.split(',')])
```

```python
# Make unique weaknesses with IDs
weaknesses_df = pd.DataFrame({
    "weakness_id": range(1, len(set(all_weaknesses)) + 1),
    "weakness_name": list(set(all_weaknesses))
})
```

```python
weaknesses_df
```

### Skills

```python
# Get all skill column names
skill_columns = [col for col in df.columns if col.startswith("tech_self_score")]

# Clean them up to be just skill names
skill_names = [col.replace("tech_self_score_", "") for col in skill_columns]

# Create a skills table
skills_df = pd.DataFrame({
    "skill_id": range(1, len(skills_df) + 1),
    "skill_name": skill_names 
})

# Re

print(skills_df)
```
